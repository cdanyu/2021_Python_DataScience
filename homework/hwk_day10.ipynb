```python
import pandas as pd
```

```python
#1. 讀取
stock_data_09 = pd.read_csv('STOCK_DAY_0050_202009.csv')
print(stock_data_09)
```

```python
stock_data_10 = pd.read_csv('STOCK_DAY_0050_202010.csv')
print(stock_data_10)
```

```python
#2. 串聯
stock_merge = pd.merge(stock_data_09, stock_data_10, on = 'date', how = 'outer')
print('merge: \n', stock_merge)
```

```python
stock_concat = pd.concat([stock_data_09, stock_data_10], axis = 0, join = 'inner')
print(stock_concat)
```

```python
stock_09_index = stock_data_09.set_index('date')
stock_10_index = stock_data_10.set_index('date')
stock_index = stock_09_index.join(stock_10_index, how = 'outer', lsuffix = '_left', rsuffix = '_right')
print('join: \n', stock_index)
```

```python
stock_append = stock_data_09.append(stock_data_10)
print('append: \n', stock_append)

stock_append_resetindex = stock_append.reset_index()
print(stock_append_resetindex)
```

```python
#3. 找出 open 小於 close 的
stock_append_resetindex.loc[stock_append_resetindex.open < stock_append_resetindex.close]
```
