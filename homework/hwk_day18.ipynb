```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 180)
y = np.cos(x * np.pi / 180)

plt.plot(x, y)
plt.show
```

```python
n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
T = np.arctan2(Y, X)

plt.scatter(X, Y, c = T, cmap = 'plasma', alpha = 0.3)

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.title('ariel')

plt.show()
```
