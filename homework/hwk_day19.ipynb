```python
n = 12
x1 = np.arange(n)

y1 = (1 - x1 / float(n)) * np.random.uniform(0.5, 1.0, n)
y2 = (1 - x1 / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(x1, +y1, facecolor = '#9999ff', edgecolor = 'white')

for i, j in zip(x1, y1):
    plt.text(i + 0.1, j + 0.05, '%.2f' % j, ha = 'center', va = 'bottom')
    
plt.bar(x1, -y2, facecolor = '#fcbea7', edgecolor = 'white')

for i, j in zip(x1, y2):
    plt.text(i, -1 * j - 0.15, '%.2f' % j, ha = 'center', va = 'bottom')

plt.ylim(-1.25, +1.25)
plt.show()
```
